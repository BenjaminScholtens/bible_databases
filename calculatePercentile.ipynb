```python
import json
import os
import numpy as np
import pandas as pd

def calculate_percentile(df):
    df['cosine_similarity'] = df['cosine_similarity'].astype(float)
    df['average_cosine_similarity'] = df.groupby('verse_uid')['cosine_similarity'].transform('mean')
    return df

# specify the directory where the json files are located
directory = './web/my-project/public/booksOfTheBibleRelationData'  # replace with your directory

# loop through all the json files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        with open(os.path.join(directory, filename), 'r') as f:
            data = json.load(f)
            df = pd.DataFrame(data)

            df = calculate_percentile(df)

            # calculate the percentile rank of the average cosine similarity across all verses
            df['cosine_similarity_percentile'] = df['average_cosine_similarity'].rank(pct=True)

            print(df)

            # If you want to save the dataframe to a new json file
            df.to_json(os.path.join(directory, f'percentile_{filename}'), orient='records')
```
